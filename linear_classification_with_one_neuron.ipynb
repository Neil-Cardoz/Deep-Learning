```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
num_points = 100
X = np.random.randn(num_points, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

```

```python
X = np.hstack((np.ones((num_points, 1)), X))  # Add bias term
```

```python
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

```python
learning_rate = 0.1
num_epochs = 1000
weights = np.zeros(X.shape[1])
```

```python
for epoch in range(num_epochs):
    for i in range(num_points):
        # Initialize weights if not already done
        z = np.dot(X[i], weights)
        # Predict class using step function
        y_pred = step_function(z)
        # Update weights
        weights += learning_rate * (y[i] - y_pred) * X[i]
```

```python
def decision_boundary(x):
    return -(weights[0] + weights[1] * x) / weights[2]
```

```python
plt.scatter(X[:, 1], X[:, 2], c=y, cmap='bwr', alpha=0.5)
x_values = np.linspace(-3, 3, 100)
y_values = decision_boundary(x_values)
plt.plot(x_values, y_values, color='black', label='Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2') 
plt.title('Linear Classification with One Neuron')
plt.legend()
plt.grid()
plt.show()
```
